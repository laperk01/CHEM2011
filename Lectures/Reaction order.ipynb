# First order
For a reaction A forming product P with rate constant *k*: $$ A \rightarrow P $$

Rate law is:
$$ R = k[A] = -\frac{d[A]}{dt} $$

and integrated form is:
$$ [A] = [A]_0 e^{-kt} $$

```python
# First create some data:
t = np.arange(0, 5, 0.1) # numpy creates an array range that is sampled evenly
A0 = 1.0 
k = 1
```

```python
# Define our first order function:
def first_order(k):
    A = A0*np.exp(-k*t)
    return A
```

```python
# Plot data:
plt.xlabel('Time /s', size=14)
plt.ylabel('Concentration of A /M', size=14)
plt.plot(t, first_order(k));
```

To check it is first order the plot of $ln[A]$ against $t$ should be linear:

```python
# Convert [A] to ln[A]:
A = first_order(k)
lnA = np.log(A) # numpy uses np.log for ln
```

```python
# Plot data:
plt.xlabel('Time /s', size=14)
plt.ylabel('Concentration of A /M', size=14)
plt.plot(t, lnA);
```

# Interactive plot

The plot below can be updated by changing the value of $k$.

```python
def plot_first_order(k):
    plt.figure()
    A = A0*np.exp(-k*t)
    plt.plot(t, A)
    plt.ylim(0, 1)
    plt.xlabel('Time /s', size=14)
    plt.ylabel('Concentration of A /M', size=14)
    plt.show()
```

```python
interactive_plot = interactive(plot_first_order, k=[0.5, 1, 3, 10])
interactive_plot
```
