# Data cleanup
Each time series file contains a recording of absorbance next to a time stamp. These should be cleaned up before analysing further. For each decay curve:
1. Delete the data before and after the time range in which the absorbance decay is smooth
2. Subtract the time value of the first data point from all of the time-data so that the decay starts at t = 0
3. Model the data using the pseudo first-order equation $ A = A_0 e^{-kt} + B $, where $B$ is an offset which accounts for the absorbance not reaching zero, or dipping below zero

```python
GA1['absorbance'].idxmax()
```

```python
GA1.head(15)
```

```python
def reset_t(f):
    """Subtract the time value of the first data point from all of the time data so decay starts at t=0"""
    
    i = f['absorbance'].idxmax() # index of maximum absorbance
    t = f['time'][i] # time at maximum absorbance
    
    ### create new data starting from top of decay curve (removing unwanted initial data)
    f_copy = f[i:].copy() # makes a copy of the data rather than overwriting
    f_copy = f_copy.reset_index(drop=True) # reset index 
    f_copy.time = f_copy.time - t # subtract time value to start decay at t=0
    
    return f_copy
```

```python
def plot_f(f):
    fig = plt.figure()
    plt.plot(f[0]['time'], f[0]['absorbance'], label='1')
    plt.plot(f[1]['time'], f[1]['absorbance'], label='2')
    plt.plot(f[2]['time'], f[2]['absorbance'], label='3')

    plt.xlabel('Elapsed time /s')
    plt.ylabel('Absorbance')
    plt.legend()
    plt.show()
```

```python
def reset_t2(f):
    """Subtract the time value of the first data point from all of the time data so decay starts at t=0
    2020 Data not using max, but data from >10s seems OK, except for HE1, where >20s 
    """
    
    t = f.query('time > 10').iloc[0][0] # time at maximum absorbance, after 10s
    i = f[f['time'] > 10].index[0]
    
    ### create new data starting from top of decay curve (removing unwanted initial data)
    reset_f = f[i:].copy() # makes a copy of the data rather than overwriting
    reset_f = reset_f.reset_index(drop=True) # reset index 
    reset_f.time = reset_f.time - t # subtract time value to start decay at t=0
    
    return reset_f
```

## Fit

Plot your decay curves and the fitted models as $(A–B) / A_0$ vs time. You may plot several on the same graph.

```python
### Fit to first order:
def first_order(t, A0, b, k):
     return A0*np.exp(-k*t)+b
```

```python
### For each series:
A = []  # Create empty list for rate constants to be added to later
B = []
C = []
D = []
E = []
```

```python
def normalise(f, optP):
    """Remove b and divide by A0"""
    
    A0 = optP[0]
    b = optP[1]
    
    return (first_order(f['time'], *optP)-b)/A0
```

```python
def append_k(f, k):
### Append k to appropriate list
    if str(f) in str(Aseries):
        A.append(k)
    
    elif str(f) in str(Bseries):
        B.append(k)

    elif str(f) in str(Cseries):
        C.append(k)

    elif str(f) in str(Dseries):
        D.append(k)

    elif str(f) in str(Eseries):
        E.append(k)
        
    else:
        print("File doesn't match?")
```

```python
def plot_fits(f_list):
    colours = ['navy', 'b', 'g', 'lime', 'orange', 'gold']
    fig = plt.figure()
    plt.xlabel('Time /s')
    plt.ylabel('(A-B)/A0')

    counter=1
    j=0
    
    for f in f_list:
        optP, pcov = opt.curve_fit(first_order, f['time'], f['absorbance'])
        A0 = optP[0]
        b = optP[1]
        k = optP[2]

        append_k(f,k)
        
        print('A0 =', A0,'+/-',2*pcov[0][0]**0.5)
        print('b =', b,'+/-',2*pcov[1][1]**0.5)
        print('k =', k,'+/-',2*pcov[2][2]**0.5,'s-1')
        print('------------------------------------')
        
        # Plotting data and fit
        plt.plot(f['time'], f['absorbance'], '.', c=colours[j], label=('exp '+str(counter)))
        plt.plot(f['time'], first_order(f['time'], *optP), "-", c=colours[j+1], label=('fit '+str(counter)))

        counter += 1
        j+= 2
        
    plt.legend()
    plt.show()
```
